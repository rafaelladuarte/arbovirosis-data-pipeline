```python
from pyspark.sql import SparkSession
from delta import configure_spark_with_delta_pip
from delta.tables import DeltaTable
import os
```

## Validar Ingestão Dengue

```python
builder = SparkSession.builder \
    .appName("exploracao-dengue") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")
```

```python
spark = configure_spark_with_delta_pip(builder).getOrCreate()
```

```python
DELTA_PATH = "/home/jovyan/work/data/bronze/dengue/delta"
```

```python
df = spark.read.format("delta").load(DELTA_PATH)

df.show(5)
```

```python
df.select(df.columns[:20]).show(5)
```

```python
df.select("ID_AGRAVO", "DT_NOTIFIC", "NU_ANO").show(5)
```

```python
df.createOrReplaceTempView("dengue23")
```

```python
spark.sql("SELECT count(*) FROM dengue23").show()
```
